```python
# Standard library imports
import json

# Third-party imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Plot style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
# Import display function for Jupyter Notebook
from IPython.display import display
from pathlib import Path

# Load `top_anime.json` from `data/raw`
data_path = Path("../data/raw/top_anime.json")
with data_path.open(encoding="utf-8") as f:
    anime_data = json.load(f)

# Convert list of dict to DataFrame
df = pd.json_normalize(anime_data)

# Show basic info and preview
df.info()
display(df.head())
```

```python
# Summary statistics for numeric columns
df.describe()
```

```python
# Check missing values
df.isna().sum().sort_values(ascending=False)
```

```python
# Plot distribution of anime scores
sns.histplot(data=df.dropna(), x="score", bins=20, kde=True)
plt.title("Distribution of Anime Scores")
plt.xlabel("Score")
plt.ylabel("Count")
plt.show()
```

```python
# Select relevant columns only
columns_to_keep = [
    "title", "type", "source", "episodes", "status", "rating",
    "score", "scored_by", "rank", "popularity", "members", "favorites",
    "year", "season", "genres", "demographics"
]

df_selected = df[columns_to_keep]
df_selected.head()
```

```python
df_selected.isna().sum().sort_values(ascending=False)
```

```python
# Fill missing values

# For numerical columns, use -1 to represent unknown
df_selected["year"].fillna(-1, inplace=True)
df_selected["episodes"].fillna(-1, inplace=True)
df_selected["rank"].fillna(-1, inplace=True)

# For categorical column, use "Unknown"
df_selected["season"].fillna("Unknown", inplace=True)

# Fill remaining missing numeric columns with -1 or 0 if appropriate
num_cols = ["score", "scored_by", "popularity", "members", "favorites"]
df_selected[num_cols] = df_selected[num_cols].fillna(-1)
```

```python
# Convert list of genres/demographics to comma-separated strings

def list_to_names_str(items):
    if isinstance(items, list):
        return ", ".join(i.get("name", "") for i in items if "name" in i)
    return ""

df_selected["genres"] = df_selected["genres"].apply(list_to_names_str)
df_selected["demographics"] = df_selected["demographics"].apply(list_to_names_str)
```

```python
# Visualize missing data
print("Duplicates:", df_selected.duplicated().sum())
```

```python
# Remove exact duplicates
df_selected = df_selected.drop_duplicates(subset=["title"])

print("Remaining rows after dropping duplicates:", len(df_selected))
```

```python
# Convert specific columns to integer type
# This is necessary for columns that should be numeric but are currently object type due to NaNs
int_cols = ["year", "rank", "episodes"]
df_selected[int_cols] = df_selected[int_cols].astype(int)
```

```python
# Display unique values for categorical columns
print(df_selected["type"].unique())
print(df_selected["status"].unique())
print(df_selected["source"].unique())
print(df_selected["season"].unique())
```

```python
# Clean up string columns
df_selected["type"] = df_selected["type"].str.strip().str.title()
df_selected["season"] = df_selected["season"].str.strip().str.title()
df_selected["status"] = df_selected["status"].str.strip().str.title()
df_selected["source"] = df_selected["source"].str.strip().str.title()
df_selected["rating"] = df_selected["rating"].str.strip().str.upper()
```

```python
# Example: Remove rows with impossible scores (except -1 placeholder)
df_selected = df_selected[(df_selected["score"] == -1) | ((df_selected["score"] >= 0) & (df_selected["score"] <= 10))]

# Display the cleaned DataFrame
df_selected.reset_index(drop=True, inplace=True)

# Reorder columns if needed
desired_order = [
    "title", "type", "source", "episodes", "status", "rating",
    "score", "scored_by", "rank", "popularity", "members", "favorites",
    "year", "season", "genres", "demographics"
]
df_selected = df_selected[desired_order]

# Ensure correct dtypes
df_selected["episodes"] = df_selected["episodes"].astype(int)
df_selected["year"] = df_selected["year"].astype(int)
df_selected["rank"] = df_selected["rank"].astype(int)
```

```python
# Final check and save
display(df_selected.head())
df_selected.info()
df_selected.to_csv("../data/processed/clean_anime_250.csv", index=False)
```
