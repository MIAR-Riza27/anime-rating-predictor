```python
# Standard library imports
import json

# Third-party imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Plot style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
# Import display function for Jupyter Notebook
from IPython.display import display
from pathlib import Path

# Load `top_anime_all.json` from `data/raw`
data_path = Path("../data/raw/top_anime_all.json")
with data_path.open(encoding="utf-8") as f:
    anime_data = json.load(f)

# Convert list of dict to DataFrame
df = pd.json_normalize(anime_data)

# Show basic info and preview
df.info()
display(df.head())
```

```python
# Summary statistics for numeric columns
df.describe()
```

```python
# Check missing values
df.isna().sum().sort_values(ascending=False)
```

```python
# Plot distribution of anime scores
sns.histplot(data=df.dropna(), x="score", bins=20, kde=True)
plt.title("Distribution of Anime Scores")
plt.xlabel("Score")
plt.ylabel("Count")
plt.show()
```

```python
# Select relevant columns only
columns_to_keep = [
    "title", "type", "source", "episodes", "status", "rating",
    "score", "scored_by", "rank", "popularity", "members", "favorites", "year", "season", "genres", "demographics"
]

df_selected = df[columns_to_keep]
df_selected.head()
```

```python
df_selected.isna().sum().sort_values(ascending=False)
```

```python
# Add indicators for missing year and season, because these are important features
df_selected["has_year"] = df_selected["year"].notna().astype(int)
df_selected["has_season"] = df_selected["season"].notna().astype(int)

# Fill missing values and convert types
fill_values = {
    "year": -1,
    "season": "Unknown",
    "score": -1,
    "scored_by": 0,
    "rank": 99999,
    "episodes": 0,
    "rating": "Unknown",
    "type": "Unknown"
}
df_selected = df_selected.fillna(fill_values)

# Convert numeric columns to appropriate types
df_selected["year"] = df_selected["year"].astype(int)
df_selected["episodes"] = df_selected["episodes"].astype(int)
df_selected["rank"] = df_selected["rank"].astype(int)
```

```python
# Clean up string columns
for col in ["type", "season", "status", "source"]:
    df_selected[col] = df_selected[col].astype(str).str.strip().str.title()
df_selected["rating"] = df_selected["rating"].astype(str).str.strip().str.upper()

# Convert list of genres/demographics to comma-separated strings
def list_to_names_str(items):
    if isinstance(items, list):
        return ", ".join(i.get("name", "") for i in items if "name" in i)
    return ""
df_selected["genres"] = df_selected["genres"].apply(list_to_names_str)
df_selected["demographics"] = df_selected["demographics"].apply(list_to_names_str)
```

```python
# Remove exact duplicates
df_selected = df_selected.drop_duplicates(subset=["title"])

# Remove rows with impossible scores (except -1 placeholder)
df_selected = df_selected[(df_selected["score"] == -1) | ((df_selected["score"] >= 0) & (df_selected["score"] <= 10))]
```

```python
# Reset index after cleaning
df_selected.reset_index(drop=True, inplace=True)

# Display unique values for categorical columns
for col in ["type", "status", "source", "season"]:
    print(f"Unique values in '{col}':", df_selected[col].unique())

# Reorder columns to include indicators
desired_order = [
    "title", "type", "source", "episodes", "status", "rating",
    "score", "scored_by", "rank", "popularity", "members", "favorites",
    "year", "season", "has_year", "has_season", "genres", "demographics"
]
df_selected = df_selected[desired_order]
```

```python
# Final check and save
display(df_selected.head())
df_selected.info()
df_selected.to_csv("../data/processed/clean_anime_full.csv", index=False)
```
